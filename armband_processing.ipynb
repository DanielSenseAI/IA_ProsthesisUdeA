```python
import os
import pandas as pd

# Choose the database to analyze
database = 'ArmbandDB0'

# Get the current working directory
current_dir = os.getcwd()
print(f"Current working directory: {current_dir}")

# Define the path to the database relative to the current working directory
database_path = os.path.join(current_dir, database)
print(f"Database path: {database_path}")

# Function to find the first .xlsx file in the directory and its subdirectories
def find_first_xlsx_file(directory):
    for root, dirs, files in os.walk(directory):
        print(f"Searching in: {root}")  # Debugging print
        for dir in dirs:
            print(f"Found directory: {dir}")  # Debugging print
        for file in files:
            print(f"Found file: {file}")  # Debugging print
            if file.endswith('.xlsx'):
                print(f"Matched file: {file}")  # Debugging print
                return os.path.join(root, file)
    return None

# Find the first .xlsx file
xlsx_file_path = find_first_xlsx_file(database_path)

if xlsx_file_path:
    # Read the .xlsx file into a pandas DataFrame
    df = pd.read_excel(xlsx_file_path)
    print(f"Loaded file: {xlsx_file_path}")
    print(df.head())
else:
    print("No .xlsx file found in the directory.")
```
